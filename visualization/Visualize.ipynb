```python
import pygame
import sys
from pygame.math import Vector2

BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

'''
input file format

map_filename
height width
robot_height robot_width
start_x start_y
finish_x finish_y
path (direction angle l r)
'''


def followPath(filename):
    global SCREEN, CLOCK
    test_file = open(filename, 'r')
    WINDOW_HEIGHT, WINDOW_WIDTH = map(int, test_file.readline().split())
    grid = [[1 for i in range(WINDOW_HEIGHT)] for j in range(WINDOWS_WIDTH)]
    for i in range(WINDOW_HEIGHT):
        l = test_file.readline();
        for j in l:
            if j == '.' or j == 'G':
                grid[i][j] = 0
    start_x, start_y = map(int, test_file,readline().split())
    robot_length, robot_width = map(int, test_file,readline().split())
    finish_x, finish_y = map(int, test_file.readline().split())
    path = []
    while line in test_file:
        direction, angle, l, r = map(int, line.split())
        path.append(direction, angle l, r))
    WINDOW_HEIGHT *= 20
    WINDOW_WIDTH *= 20
    i = 0
    
    pygame.init()
    SCREEN = pygame.display.set_mode((WINDOW_HEIGHT, WINDOW_WIDTH))
    CLOCK = pygame.time.Clock()
    SCREEN.fill(WHITE)
    
    rect = pygame.Rect(start_x * 20 + 10, start_y * 20 + 10, robot_length, robot_width)
    bot_vel = 20
    l_curr = 0
    dx_curr = 0
    dy_curr = 0
    turn = false

    while True:
        pos = Vector2(rect.x, rect.y)
        dt = CLOCK.tick(60)
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
        if l_curr < path[i][2]:
            if turn:
                dy = bot_vel * dt / 1000   
                rect.move_ip(0, dy)
                l_curr += 1
            else:
                dx = bot_vel * dt / 1000   
                rect.move_ip(dx, 0)
                l_curr += 1
        else:
            if path[i][3] == -1:
                i += 1
            else:
                pos += Vector2(rect.x, rect.y).rotate(path[i][1] * direction) * dt
                turn = !turn
                
        pygame.draw.rect(SCREEN, BLUE, rect)
        drawGrid(grid)
        pygame.display.flip()


def drawGrid(grid):
    SCREEN.fill(WHITE)
    blockSize = 20
    for x in range(WINDOW_WIDTH):
        for y in range(WINDOW_HEIGHT):
            rect = pygame.Rect(x*blockSize, y*blockSize,
                               blockSize, blockSize)
            pygame.draw.rect(SCREEN, BLACK, rect, grid[x][y])
```
